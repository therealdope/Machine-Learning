# AND_NN_Backpropagation

```python
import numpy as np
import pandas
import matplotlib.pyplot as plt

data = pandas.read_csv(filepath_or_buffer='/content/drive/MyDrive/ML/P8-P9/AND.csv', header=None)
data = data.apply(pandas.to_numeric, errors='coerce')
data = data.dropna()

# Convert dataframe to numpy array
data = data.values.astype(float)


n_datapoints = data.shape[0]
n_dimensions = data.shape[1] - 1


W = 2 * np.random.random_sample((n_dimensions)) - 1
b = np.random.random()

# initial weights and bias
print("Weight: ",W, W.shape)
print("bias: ",b)

# learning rate and number of epochs
lr = 0.1
n_epochs = 50

for epoch in range(n_epochs):
    # Inner for loop for presentation of all input/output pairs in an epoch
    for i in range(n_datapoints):
        # Ensure input data is a numpy array of floats
        input_data = data[i, 0:n_dimensions]

        # net_input = Wp + b
        net_input = np.dot(W, input_data) + b

        # a = 1 if net_input >= 0 else 0 (hardlim activation function)
        a = 1 if net_input >= 0 else 0

        # error = target - actual
        e = data[i, n_dimensions] - a

        # Update weights and bias using perceptron learning rule
        W = W + lr * e * input_data
        b = b + lr * e

        print(f"Epoch {epoch}, Data point {i}, Weights: {W}, Bias: {b}")

# Printing final weights and bias
print("Final Weights:", W)
print("Final Bias:", b)

# Making prediction
prediction = (np.dot(data[:, 0:n_dimensions], W) + b) >= 0

# Displaying prediction
print("Predictions:", prediction.astype(int))
```
