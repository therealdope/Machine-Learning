```python
#importing libraries
import pytesseract

import os #for the file path

from PIL import Image #for open the image

from google.colab.patches import cv2_imshow #for display the image

import cv2 #for read the image

from matplotlib import pyplot as plt #for display the image
```

```python
#filepath
fp = input("Enter the file path: ")

#image recognition
img = cv2.imread(fp)
img1 = Image.open(fp)

plt.imshow(img1)
cv2_imshow(img)

info = pytesseract.image_to_string(img)
print(info)
```
